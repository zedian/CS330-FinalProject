```python
import pickle, itertools
import numpy as N
import pandas
import matplotlib.pyplot as pyplot
import seaborn
import language_models
import utils

paths = language_models.get_paths('bert')
```

```python
def grid_share_axes(axs, shareX=True, shareY=True):
    nRow = len(axs)
    nCol = len(axs[0])
    if shareX:
        [axs[0][j].get_shared_x_axes().join(*[axs[i][j] for i in range(nRow)])
         for j in range(nCol)]
        # Remove ticks
        [axs[i][j].set_xticklabels([])
         for i in range(nRow-1)
         for j in range(nCol)]
    if shareY:
        [axs[i][0].get_shared_y_axes().join(*[axs[i][j] for j in range(nCol)])
         for i in range(nRow)]
        [axs[i][j].set_yticklabels([])
         for i in range(nRow)
         for j in range(1,nCol)]
def label_right(axs_row, label):
    """
    Use in conjunction as the above
    """
    ax = axs_row[-1]
    ax.set_ylabel(label)
    ax.yaxis.set_label_position("right")
    ax.yaxis.tick_right()
    
```

## Cosine Gradient Plot

```python
modelTypes = ['tiny', 'mini', 'small', 'medium']

gradCosines = dict()
for mt in modelTypes:
    with open(paths['plots'] / f'{mt}-gradCosine.pickle', 'rb') as f:
        gradCosines[mt] = pickle.load(f)

# Print example keys
print(gradCosines['tiny'][1])
```

### Cosine Rankings (averaged across history)

```python
fig, axs = pyplot.subplots(2, 4, figsize=(16,6))

def to_layer(name, score):
    name = name.split('.')
    if name[2] == 'encoder':
        return f'{name[2]}.{name[4]}'
    if name[2] == 'decoder':
        return f'{name[2]}.{name[4]}'
    
    return name[2]
def to_color(score, alternate):
    if score > 0:
        return 'teal' if alternate % 2 == 0 else 'steelblue'
    else:
        return 'crimson' if alternate % 2 == 0 else 'chocolate'

for i,mt in enumerate(modelTypes):
    gc, keys = gradCosines[mt]
    scores = gc.mean(axis=0)
    df = pandas.DataFrame({'Rank': range(1,scores.size+1),
                           'Sorted': sorted(list(scores), reverse=True),
                           'Unsorted': scores,
                           'Layer': [to_layer(n,s) for n,s in zip(keys,scores)]})
    layerLookup = [k for k, _group in itertools.groupby(df['Layer'])]
    layerLookup = {k:i for i,k in enumerate(layerLookup)}
    df['Alternate'] = [layerLookup[l] for l in df['Layer']]
    df['Color'] = [to_color(s, a) for s,a in zip(df['Unsorted'], df['Alternate'])]

    ax = axs[0][i]
    ax.bar(x=df['Rank'], height=df['Sorted'],
           color=['teal' if c > 0 else 'crimson' for c in df['Sorted']])
    ax.set_ylim(-0.025,0.17)
    ax.set_title(mt)
    
    ax = axs[1][i]
    ax.bar(x=df['Rank'], height=df['Unsorted'], color=df['Color'])
    ax.set_ylim(-0.025,0.17)

grid_share_axes(axs)
label_right(axs[0], 'Ranking')
label_right(axs[1], 'Order')

fig.tight_layout()
artist_suptitle = fig.suptitle(f'Gradient Cosine Rankings for models', fontsize=18, y=1.03)
fig.savefig(paths['plots'] / 'bert-gradient-cosine.png', dpi=300, bbox_inches='tight',bbox_extra_artists=[artist_suptitle])
```

### Historical trends of the highest and lowest cosines

```python
# plots
fig, axs = pyplot.subplots(2, len(modelTypes), figsize=(4*len(modelTypes), 6))

heuristic_results = dict()

def format_key(k):
    k = k.split('.')
    k = k[2:]
    return '.'.join(k)

for i, mt in enumerate(modelTypes):
    gc, keys = gradCosines[mt]
    
    # Find highest and lowest cosine key
    
    scores = gc.mean(axis=0)
    jMin, jMax = scores.argmin(), scores.argmax()
    
    # Plot the historic trends
    ax = axs[0][i]
    ax.axhline(y=0, xmin=0.0, xmax=1.0, color='k', linewidth=1)
    ax.plot(gc[:,jMax], color='teal', alpha=0.2)
    ax.plot(utils.smooth(gc[:,jMax], filter_size=10), color='teal')
    ax.set_ylim([-1.1, 1.1])
    ax.set_title(mt)
    #ax.set_title(format_key(keys[jMax]))
    
    ax = axs[1][i]
    ax.set_xlabel('Step')
    ax.axhline(y=0, xmin=0.0, xmax=1.0, color='k', linewidth=1)
    ax.plot(gc[:,jMin], color='crimson', alpha=0.2)
    ax.plot(utils.smooth(gc[:,jMin], filter_size=10), color='crimson')
    ax.set_ylim([-1.1, 1.1])
    #ax.set_title(format_key(keys[jMin]))
    

grid_share_axes(axs)
label_right(axs[0], 'Highest')
label_right(axs[1], 'Lowest')

fig.tight_layout()
artist_suptitle = fig.suptitle(f'Historical Gradient Cosines for lowest and highest parameters', fontsize=18, y=1.03)
fig.savefig(paths['plots'] / 'bert-gradient-cosine-hist.png', dpi=300, bbox_inches='tight',bbox_extra_artists=[artist_suptitle])
```

## Training Plots

```python

def plot_single_model(source):
    TOPOLOGIES = [
        ('shared', 'paleturquoise'),
        ('separate', 'lightcoral'),
        ('surgical', 'grey')
    ]
    TASKS = ['mrpc', 'sst2']
    losses = dict()
    metrics = dict()
    eval_losses = dict()
    eval_metrics = dict()
    
    def transform_input(x):
        # Convert list of (step, {}) -> {k: [ ... ]}
        # effectively a transposition
        keys = x[0][1].keys()
        return {k:[(s,val[k]) for s,val in x]
                for k in keys}
    
    for top,_ in TOPOLOGIES:

        with open(paths['plots'] / f"{source}-{top}.pickle", "rb") as f:
            exp = pickle.load(f)
            losses[top] = transform_input(exp[0])
            metrics[top] = transform_input(exp[1])
            eval_losses[top] = transform_input(exp[2])
            eval_metrics[top] = transform_input(exp[3])
            
    
    fig, axs = pyplot.subplots(len(TASKS), 4, figsize=(16, len(TASKS)*3))
    for iTask,task in enumerate(TASKS):
        for iTop,(top,color) in enumerate(TOPOLOGIES):
        
            # plot
            ax = axs[iTask][0]
            tl_x, tl_y = zip(*losses[top][task])
            ax.plot(tl_x, tl_y, label=f'{top}', color=color)
            

            ax = axs[iTask][1]
            el_x, el_y = zip(*eval_losses[top][task])
            ax.plot(el_x, el_y, label=f'{top}', color=color)
            

            ax = axs[iTask][2]
            tm_x, tm_y = zip(*metrics[top][task])
            ax.plot(tm_x, tm_y, label=f'{top}', color=color)
            #ax.set_ylim([0.95, 0.999])
            

            ax = axs[iTask][3]
            em_x, em_y = zip(*eval_metrics[top][task])
            ax.plot(em_x, em_y, label=f'{top}', color=color)
            #ax.set_ylim([0.95, 0.999])
            
        label_right(axs[iTask], task)
        
    for i in range(4):
        ax = axs[0][i]
        ax.legend()
        ax = axs[1][i]
        ax.set_xlabel('Step')
        
    ax = axs[0][0]
    ax.set_title('Train Loss')
    ax.set_yscale('log')
    ax = axs[0][1]
    ax.set_title('Val Loss')
    ax.set_yscale('log')
    ax = axs[0][2]
    ax.set_title('Train Accuracy')
    ax = axs[0][3]
    ax.set_title('Val Accuracy')
    
    grid_share_axes(axs, shareX=True, shareY=False)
    
    fig.tight_layout()
    artist_suptitle = fig.suptitle(f'bert-{source} Train history', fontsize=18, y=1.03)
    fig.savefig(paths['plots'] / f'bert-{source}-hist.png', dpi=300, bbox_inches='tight',bbox_extra_artists=[artist_suptitle])
    
plot_single_model('medium')
```

```python
plot_single_model('small')
plot_single_model('mini')
plot_single_model('tiny')
```
